```python
df_s9=df[df['season']=='2023/24']
```

```python
p1=df_s9_pp.groupby(['batting_team'])['runs_off_bat','extras'].sum()
```

```python
p2=df_s9_pp.groupby(['batting_team'])['ball','wides','noballs'].count()
```

```python
pp_batting=p1.merge(p2,on='batting_team')
```

```python
pp_batting['runs']=pp_batting['runs_off_bat']+pp_batting['extras']
```

```python
pp_batting['balls']=pp_batting['ball']-pp_batting['wides']-pp_batting['noballs']
```

```python
pp_batting['avg_run_rate']=pp_batting['runs']/pp_batting['balls']*6
```

```python
pp_batting['wickets_lost']=df_s9_pp.groupby(['batting_team'])['player_dismissed'].count()
```

```python
pp_batting['average']=pp_batting['runs']/pp_batting['wickets_lost']
```

```python
pp_batting.sort_values(by='avg_run_rate',ascending=False)
```

```python
pp_batting.sort_values(by='average',ascending=False)
```
